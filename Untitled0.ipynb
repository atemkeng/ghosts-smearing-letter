```python
import numpy as np
import pylab
%matplotlib inline


pylab.rc('xtick', labelsize=14.5) 
pylab.rc('ytick', labelsize=14.5) 

params = {'legend.fontsize': 18, 
          'legend.linewidth': 2} 
pylab.rcParams.update(params) 
  
textlabel = 18
```

```python
sf_l15arcmin_2MHZ= np.load("sf-L15arcmin-BW2MHz-T1s.npy")
sf_l15arcmin_1MHZ= np.load("sf-L15arcmin-BW1MHz-T1s.npy")
sf_l15arcmin_3MHZ= np.load("sf-L15arcmin-BW3MHz-T1s.npy")
sf_l15arcmin_4MHZ= np.load("sf-L15arcmin-BW4MHz-T1s.npy")
sf_l15arcmin_6MHZ= np.load("sf-L15arcmin-BW6MHz-T1s.npy")
sf_l15arcmin_8MHZ= np.load("sf-L15arcmin-BW8MHz-T1s.npy")

baseline = np.load("baselineLeng.npy")

baseline = baseline/1000.

pylab.figure(figsize=(16, 10))
pylab.plot(baseline, sf_l15arcmin_1MHZ, 'k*', label='1 MHz')
pylab.plot(baseline, sf_l15arcmin_2MHZ, 'bo', label='2 MHz')
pylab.plot(baseline, sf_l15arcmin_3MHZ, 'go', label='3 MHz')
pylab.plot(baseline, sf_l15arcmin_4MHZ, 'yo', label='4 MHz')
pylab.plot(baseline, sf_l15arcmin_6MHZ, 'ko', label='6 MHz')
pylab.plot(baseline, sf_l15arcmin_8MHZ, 'ro', label='8 MHz')

print "For natural weighting the PSF peak will be:", np.mean(sf_l15arcmin_1MHZ)

pylab.title(r"VLAA at 3GHz, source at 15 arcmin",fontsize=textlabel) 
pylab.xlabel(r"East-West baseline, km",fontsize=textlabel) 
pylab.ylabel(r"Smearing factor (SF)",fontsize=textlabel) 
pylab.axhline(0.,color="0.5",ls=':')
pylab.ylim(np.min(sf_l15arcmin_6MHZ)-0.2,1.1)
pylab.xlim(0.,np.max(baseline)+1)
pylab.legend()
```
